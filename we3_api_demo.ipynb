In order to run the notebook yourself, you may need to create your own account on node provider website. Here we show you how we do it on Infura (one of the many node providers).
        


After creating the account, you have to verify the email address. Don't worry, we are only going to use Free version of the API so it won't ask for credit card etc. 



After that you can launch your own instance of Ethereum blockchain


You will need to create a key that is a verification mechanism to use the node that you launced.
